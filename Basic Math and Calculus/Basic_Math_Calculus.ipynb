```python
from sympy import *
import matplotlib.pyplot as plt
```

```python
# Declaring linear function in Python

def f(x):
  return 2 * x + 1

x_values = [0, 1, 2, 3]

for x in x_values:
  y = f(x)
  print(y)
```

```python
# charting a linear function
x = symbols('x')
f = 2 * x + 1
plot(f)
```

```python
# charting an exponential function
x = symbols('x')
f = x**2 + 1
plot(f)
```

When a function is continuous but curvy, rather than llinear and straight, we call it a curvilinear function.

```python
def f(x, y):
  return 2*x + 3*y

x_values = [1, 2, 3]
y_values = [1, 2, 3]

for x, y in zip(x_values, y_values):
 func = f(x, y)
 print(func)
```

```python
from sympy.plotting import plot3d

x, y = symbols('x y')
f = 2*x + 3*y
plot3d(f)
```

### Summations

```python
summation = sum(2 * i for i in range(1, 6))
print(summation)
```

```python
x = [1, 4, 6, 2]
n = len(x)

summation = sum(10*x[i] for i in range(0, n))
print(summation)
```

```python
i, n = symbols('i n')

# iterate each element i from 1 to n
# then multiply and sum
summation = Sum(2 * i, (i, 1, n))

# specify n as 5
# iterating the numbers 1 through 5
up_to_5 = summation.subs(n, 5)
print(up_to_5.doit())
```

### Euler's number and natural logarithms

##### Compounding interest

```python
p = 100 # investment
r = .20 # interest rate
t = 2.0 # number of years
n = 12 # number of months each years

# A = P X (1 + r / n)**nt

def calculate_interest(p, r, t, n):
  a = p * (1 + (r/n))**(n*t)
  return a
print(calculate_interest(p, r, t, n))

n = 365
print(calculate_interest(p, r, t, n))

n = 8769
print(calculate_interest(p, r, t, n))
n = 525600
print(calculate_interest(p, r, t, n))
```

##### Continuous interst

```python
from math import exp 

# A = P x e*rt

p = 100 # starting amount
r = .20 # interest rate, by year
t = 2.0 # time, number of years

a = p * exp(r * t)
print(a)
```
