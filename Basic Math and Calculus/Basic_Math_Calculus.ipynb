### Euler's number and natural logarithms

##### Compounding interest

$$
A = P  (1 + \frac{r}{n})^{nt}
$$

```python
p = 100 # investment
r = .20 # interest rate
t = 2.0 # number of years
n = 12 # number of months each years

# A = P X (1 + r / n)**nt

def calculate_interest(p, r, t, n):
  a = p * (1 + (r/n))**(n*t)
  return a
print(calculate_interest(p, r, t, n))

n = 365
print(calculate_interest(p, r, t, n))

n = 8769
print(calculate_interest(p, r, t, n))
n = 525600
print(calculate_interest(p, r, t, n))
```

### Limits

As with Euler's number, some intersting ideas emerge when we forever increase or decreaase an input variable and the output variable keeps approaching a value but never reaching it.

$$
f(x) = \frac{1}{x}\
$$

as x increases forever, f(x) gets closer to 0, but never reaches it. The way we express a value that is forever being approached, but never reached, is through a limit:

$$
\lim_{{x \to \infty}}\frac{1}{x} = 0\
$$
The way we read this is ''as x approaches infinity'', the function $$\frac{1}{x}$$ approaches 0, but never reaches 0

```python
x = symbols('x')
f = 1 / x
p = plot(f, (x, 0, 10), ylim=(0, 10), show=False, legend=True)
p[0].line_color = 'blue'
p[0].label = 'f(x) = 1/x'
p.show()
```

```python
from sympy import *

x = symbols('x')
f = 1 / x
result = limit(f, x, oo)
print(result)
```

### Derivatives

A <i>derivative</i> tells the slope of a function, and it is useeful to measure the rate of change at any point in a function. Derivatives are often used in machine learning and other mathematical algorithms, especially gradient descent. When the slope is 0, that means we are at the minmum or maximum of an output variable.

Let's start with the example:
$$
f(x) = x^2
$$

how steep is the curve at x = 2 and how steep at 2.1

1. Calculate the derivative of f(x):
$$f'(x) = \frac{d}{dx}(x^2) = 2x$$
2. Find the slope m at x = 2
$$m = f'(2) = 2.2 = 4$$
3. Now we have the point (2, 4) and the slop m = 4
4. Plug these values into the point-slope form:
$$y - 4 = 4(x - 2)$$
5. Simplify and express the equation in slope-intercept form (if desired):
$$y=4x-4$$

So, the equation of the tangent line to $$f(x) = x^2$$ at  $$x = 2$$ is $$y = 4x - 4$$

```python
# Define the symbol
def calculate_and_plot_slope(value):
  x = symbols('x')
  # Define the function f(x)
  f = x**2
  # Calculate the derivative of f(x)
  f_prime = diff(f, x)
  print(f_prime)
  # the point where to find the tangent line
  x_value = value
  # calculate the slope of the tangent line at a specified point
  slope = f_prime.subs(x, x_value)
  # calculate the equation of the tangent line using the point-slope form
  tangent_line = slope * (x - x_value) + f.subs(x, x_value)
  print(tangent_line)
  print(slope)

  # create and extend the plot with the tangent line
  p = plot(f, xlim=(-5, 5), ylim=(0, 15), show=False, legend=True)
  p.extend(plot(tangent_line, show=False, line_color='red', legend=True))
  p.show()
```

```python
calculate_and_plot_slope(2)
```

To find the derivative you can use the **Power Rule fore differentiation**
$$nx^{(n-1)}$$

```python
def derivative_x(f, x, step_size):
  m = (f(x + step_size) - f(x)) / ((x + step_size) - x)
  return m

def my_function(x):
  return x**2

slope_at_2 = derivative_x(my_function, 2, .00001)
print(slope_at_2)
```

```python
x = symbols('x')
f = x**2
f_prime = diff(f, x)
x_value = 2
slope = f_prime.subs(x, x_value)
print(slope)
```

### Partial Derivatives

Partial derivatives are derivatives of functions that have multiple input variables. Rather than finding the slope on a one-dimensional function, we have slopes with respect to multiple variables in several directions.

$$f(x, y) = 2x^3 + 3y^3$$
the x and y variable each get their own derivatives: $$\frac{d}{dx}$$ and $$\frac{d}{dy}$$
Thse represent the slope values with respect to each variable on a multidimensional surface. We technically call these 'slopes' gradients when dealing with multiple dimensions.

So the derivative of $$2x^3 = 2\frac{d}{dx} = 2(nx^{n-1}) = 2(3x)^2 = 6x^2$$
and the derivative of $$3y^3 = 3\frac{d}{dy} = 3(ny^{n-1}) = 3(3y)^2 = 9y^2$$

```python
from sympy import *
from sympy.plotting import plot3d

x, y = symbols('x, y')

f = 2*x**3 + 3*y**3

dx_f = diff(f, x)
dy_f = diff(f, y)

print(dx_f)
print(dy_f)

plot3d(f)
```
