### Limits

As with Euler's number, some intersting ideas emerge when we forever increase or decreaase an input variable and the output variable keeps approaching a value but never reaching it.

$$
f(x) = \frac{1}{x}\
$$

as x increases forever, f(x) gets closer to 0, but never reaches it. The way we express a value that is forever being approached, but never reached, is through a limit:

$$
\lim_{{x \to \infty}}\frac{1}{x} = 0\
$$
The way we read this is ''as x approaches infinity'', the function $$\frac{1}{x}$$ approaches 0, but never reaches 0

```python
x = symbols('x')
f = 1 / x
p = plot(f, (x, 0, 10), ylim=(0, 10), show=False, legend=True)
p[0].line_color = 'blue'
p[0].label = 'f(x) = 1/x'
p.show()
```

### Derivatives

A <i>derivative</i> tells the slope of a function, and it is useeful to measure the rate of change at any point in a function. Derivatives are often used in machine learning and other mathematical algorithms, especially gradient descent. When the slope is 0, that means we are at the minmum or maximum of an output variable.

Let's start with the example:
$$
f(x) = x^2
$$

how steep is the curve at x = 2 and how steep at 2.1

1. Calculate the derivative of f(x):
$$f'(x) = \frac{d}{dx}(x^2) = 2x$$
2. Find the slope m at x = 2
$$m = f'(2) = 2.2 = 4$$
3. Now we have the point (2, 4) and the slop m = 4
4. Plug these values into the point-slope form:
$$y - 4 = 4(x - 2)$$
5. Simplify and express the equation in slope-intercept form (if desired):
$$y=4x-4$$

So, the equation of the tangent line to $$f(x) = x^2$$ at  $$x = 2$$ is $$y = 4x - 4$$

```python
# Define the symbol
def calculate_and_plot_slope(value):
  x = symbols('x')
  # Define the function f(x)
  f = x**2
  # Calculate the derivative of f(x)
  f_prime = diff(f, x)
  print(f_prime)
  # the point where to find the tangent line
  x_value = value
  # calculate the slope of the tangent line at a specified point
  slope = f_prime.subs(x, x_value)
  # calculate the equation of the tangent line using the point-slope form
  tangent_line = slope * (x - x_value) + f.subs(x, x_value)
  print(tangent_line)
  print(slope)

  # create and extend the plot with the tangent line
  p = plot(f, xlim=(-5, 5), ylim=(0, 15), show=False, legend=True)
  p.extend(plot(tangent_line, show=False, line_color='red', legend=True))
  p.show()
```

```python
calculate_and_plot_slope(2)
```

To find the derivative you can use the **Power Rule fore differentiation**
$$nx^{(n-1)}$$

```python
def derivative_x(f, x, step_size):
  m = (f(x + step_size) - f(x)) / ((x + step_size) - x)
  return m

def my_function(x):
  return x**2

slope_at_2 = derivative_x(my_function, 2, .00001)
print(slope_at_2)
```

### Partial Derivatives

Partial derivatives are derivatives of functions that have multiple input variables. Rather than finding the slope on a one-dimensional function, we have slopes with respect to multiple variables in several directions.

$$f(x, y) = 2x^3 + 3y^3$$
the x and y variable each get their own derivatives: $$\frac{d}{dx}$$ and $$\frac{d}{dy}$$
Thse represent the slope values with respect to each variable on a multidimensional surface. We technically call these 'slopes' gradients when dealing with multiple dimensions.

So the derivative of $$2x^3 = 2\frac{d}{dx} = 2(nx^{n-1}) = 2(3x)^2 = 6x^2$$
and the derivative of $$3y^3 = 3\frac{d}{dy} = 3(ny^{n-1}) = 3(3y)^2 = 9y^2$$

```python
from sympy import *
from sympy.plotting import plot3d

x, y = symbols('x, y')

f = 2*x**3 + 3*y**3

dx_f = diff(f, x)
dy_f = diff(f, y)

print(dx_f)
print(dy_f)

plot3d(f)
```

### Using limits to calculate  derivatives

```python
from sympy import *

x, s = symbols('x s')

# declare function
f = x ** 2

# slope between two points with gap 's'
# substitute into rise over run formula
slope_f = (f.subs(x, x + s) - f) / ((x + s) - x)

# substitue 2 for x
slope_2 = slope_f.subs(x, 2)

# calculate slope at x = 2
# infinitely approach step size _s_ to 0
result = limit(slope_2, s, 0)

print(result)
```

### The Chain Rule 

$$y = x^2 + 1$$
$$z = y^3 - 2$$
$$z = (x^2 + 1)^3 - 2$$
derivative: $$\frac{dz}{dx}$$
power rule: $$nz^{n-1}$$
$$u(z) = 3(x^2 + 1)^2$$
$$v(x) = 2x$$
$$z'(x) = 3(x^2 + 1)^2 * 2x$$
$$z'(x) = 6x(x^2 + 1)^2$$

You can also take the derivate of the y and z functoins separately, and then multiple them together, this also produces the derivative of z with respect to x

$$\frac{dy}{dx}(x^2 + 1) = 2x$$
$$\frac{dz}{dy}(y^3 - 2) = 3y^2$$
$$\frac{dz}{dx} = (2x)(3y^2) = 6xy^2$$
substitutin y:  $$6x(x^2 + 1)^2$$


```python
from sympy import *

z = (x**2 + 1)**3 - 2
dz_dx = diff(z, x)
print(dz_dx)
```

```python
# Calculating the derivative dz/dx with and without the chain rule

x, y = symbols('x y')

# derive for first function
# need to underscore y to prevent variable crash
_y = x**2 + 1
dy_dx = diff(_y)

# derivative for second function
z = y**3 - 2
dz_dy = diff(z)

# calculate derivative with and without
# chain rule, substitue y function
dz_dx_chain = (dy_dx * dz_dy).subs(y, _y)
dz_dx_no_chain = diff(z.subs(y, _y))

# Prove chain rule by showing both are equal
print(dz_dx_chain) 
print(dz_dx_no_chain)
```

### Integrals

The opposite of a derivative is an <i>integral</i>, which finds the area under the curve for a given range.

```python
def approximate_integral(a, b, n, f):
  delta_x = (b - a) / n 
  total_sum = 0

  for i in range(1, n + 1):
    midpoint = 0.5 * (2 * a + delta_x * (2 * i - 1))
    total_sum += f(midpoint)

  return total_sum * delta_x

def my_function(x):
  return x**2 + 1

area = approximate_integral(a=0, b=1, n=5, f=my_function)
print(area)
area = approximate_integral(a=0, b=1, n=1000, f=my_function)
print(area)
area = approximate_integral(a=0, b=1, n=1000000, f=my_function)
print(area)
```

### Calculating Integrals with Limits

```python
# Declare variables
x, i, n = symbols('x i n')

# declare function and range
f = x**2 + 1
lower, upper = 0, 1

# calculate width and each reactagnel height at index 'i'
delta_x = ((upper - lower) / n)
x_i = (lower + delta_x * i)
fx_i = f.subs(x, x_i)

# Iterate all 'n' rectangles and sum their areas
n_rectangles = Sum(delta_x * fx_i, (i, 1, n)).doit()

# Calculate the area by approaching the number 
# of rectangles 'n' to infinity
area = limit(n_rectangles, n, oo)
print(area)
```
